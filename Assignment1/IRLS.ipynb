```python
import IRLS

x_arr,y_arr=IRLS.loadDotData("a9a/a9a")
x_test,y_test=IRLS.loadDotData("a9a/a9a.t")


trainHistory=IRLS.train(x_arr,y_arr,x_test,y_test)
IRLS.test(x_test,y_test,regularizer="")
```

```python
IRLS.plotLossAcc(trainHistory,"IRLS on a9a data\n W^2=38.5209 AUC=0.9014")
```

```python
trainHistoryL2=IRLS.train(x_arr,y_arr,x_test,y_test,regularizer="L2",w_lambda= 0.3)
IRLS.test(x_test,y_test,regularizer="L2")
```

```python
IRLS.plotLossAcc(trainHistoryL2,"IRLS L2-norm on a9a data \n W^2=7.1097  AUC=0.9020")
```
