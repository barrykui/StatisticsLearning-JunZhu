```python
import numpy as np
import matplotlib.pyplot as plt
def plotloss(logfile, title): 
    with open(logfile,'rb') as f:
        #f.readline()
        lines = f.readlines()
        #random.shuffle(lines)
        data = {"train_time":[],"train_loss":[],"train_error":[],"test_time":[],"test_loss":[],"test_error":[]}
        best_loss = []
        
        test_acc = []
        best_iter = []
        best_iter.append(0)
        test_l2norm = []
        lambdalist=[]
        lambdalist.append(-0.01)
        d_targ = {}# copy.deepcopy(initDict())
        d_cor2 = {}# copy.deepcopy(initDict())
        d_pred = {}# copy.deepcopy(initDict())
        precisionClass = []
        for line in lines:
            if line[0:7]=="lambda:":
                info = line[line.index("lambda:"):].strip()
                info = info.split()
                #print(info)

                #data['train_time'].append(info[1])
                lambdalist.append(float(info[1]))

            if line[0:11]==" * testACC:":
                info = line[line.index("testACC:"):].strip()
                info = info.split()
                #print(info)

                #data['test_time'].append(info[1])
                test_acc.append(float(info[1]))
                test_l2norm.append(float(info[19]))
                
            if line[0:40]=="Stoped by earlystoping, best model loss:":
                info = line[line.index("loss:"):].strip()
                info = info.split()
                #print(info)
                #data['test_time'].append(info[1])
                best_loss.append(float(info[1]))
                best_iter.append(float(info[4]))
                

        print(len(lambdalist))
        print(len(best_iter))
        
        fig, ax = plt.subplots(1, 1, figsize=(10, 4))
        ax.bar(lambdalist,best_iter,0.005,label="")
        #ax[0].plot(data['test_loss'],label="Testing Loss")
        ax.set_ylabel('Convergence Iteration')
        ax.set_xlabel('Lambda')
        title0="Convergence Iteration with different lambda"
        ax.set_title(title0)
        ax.legend(loc='upper right', shadow=True)
        fig.tight_layout()
        plt.show()
        
        fig, ax = plt.subplots(1, 1, figsize=(10, 4))
        #error
        ax.plot(lambdalist[1:],test_acc,label="")
        #ax[1].plot(data['test_error'],label="Top1 Testing Error")
        ax.set_ylabel('Test Accuracy')
        ax.set_xlabel('Lambda')
        title1="Test Accuracy with different lambda"
        ax.set_title(title1)

        ax.legend(loc='upper right', shadow=True)
        fig.tight_layout()

        plt.show() 
        
        
        fig, ax = plt.subplots(1, 1, figsize=(10, 4))
        #error
        ax.plot(lambdalist[1:],test_l2norm,label="")
        #ax[1].plot(data['test_error'],label="Top1 Testing Error")
        ax.set_ylabel('L2(W)')
        ax.set_xlabel('Lambda')
        title1="L2(W)  with different lambda"
        ax.set_title(title1)

        ax.legend(loc='upper right', shadow=True)
        fig.tight_layout()

        plt.show() 
plotloss("testLambda-all.log","")
```
